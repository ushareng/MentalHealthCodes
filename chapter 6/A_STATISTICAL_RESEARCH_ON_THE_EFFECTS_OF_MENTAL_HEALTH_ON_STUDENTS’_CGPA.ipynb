In this part we take a student mental health dataset- data  and and try to predict the effects of studentâ€™s mental health on their CGPA. This dataset was generated by a university through google form survey to examine the current academic situation and mental health of their students. This dataset consists of various physiological and mental health attributes and more than 200 datapoints.





This task can be viewed as a tabular classification task where we can leverage the tubular techniques like decision models in order to predict the CGPA of students based on their physiological and mental health condition. We use the label encoder to convert the string labels into integer classes so that it can be fed into the models for training. Then we move on to build our model for predictions. We make use of one of the best models available i.e. XGBoost model.


```python
import pandas as pd
df = pd.read_csv('/content/Student Mental health.csv')
df.head()
```

```python
X= df.drop(columns=['What is your CGPA?','Timestamp'],axis=1)
y= df['What is your CGPA?']
```

```python
from sklearn.preprocessing import LabelEncoder
X = X.apply(LabelEncoder().fit_transform)
X.head()
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from xgboost import XGBClassifier

bst = XGBClassifier()
bst.fit(X, y)
```

```python
bst.score(X,y)
```

```python
print(X.iloc[0,:])
print(' ')
print('Prediction: ', le.classes_[bst.predict(X.iloc[0,:].to_numpy().reshape(1,-1))])
```
