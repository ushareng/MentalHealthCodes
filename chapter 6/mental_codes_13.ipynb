```python
import pandas as pd
df = pd.read_csv('/content/Student Mental health.csv')
df.head()
```

```python
X= df.drop(columns=['What is your CGPA?','Timestamp'],axis=1)
y= df['What is your CGPA?']
```

```python
from sklearn.preprocessing import LabelEncoder
X = X.apply(LabelEncoder().fit_transform)
X.head()
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from xgboost import XGBClassifier

bst = XGBClassifier()
bst.fit(X, y)
```

```python
bst.score(X,y)
```
