```python
import pandas as pd
df = pd.read_csv('/content/MentalHealthDataset.csv')
df.head()
```

```python
X= df.drop(columns=['Growing_Stress'],axis=1)
y= df['Growing_Stress']
```

```python
from sklearn.preprocessing import LabelEncoder
X = X.apply(LabelEncoder().fit_transform)
X.head()
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X, y)
```

```python
clf.score(X,y)
```
